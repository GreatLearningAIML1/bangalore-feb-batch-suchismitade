### Build a DNN using Keras with `RELU` and `ADAM`

#### Load tensorflow

```python
import tensorflow as tf
```

```python
import numpy as np
import pandas as pd
from keras.datasets import mnist
from keras.utils import np_utils
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
```

#### Collect Fashion mnist data from tf.keras.datasets 

```python
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

#### Change train and test labels into one-hot vectors

```python
#Encoding the output class label (One-Hot Encoding)
train_Y=np_utils.to_categorical(y_train)
test_Y=np_utils.to_categorical(y_test)
```

```python
train_Y[0:1]
```

```python
x_train.shape
```

#### Build the Graph

#### Initialize model, reshape & normalize data

```python
#Initialize model, reshape & normalize data
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Reshape((784,),input_shape=(28,28,)))
model.add(tf.keras.layers.BatchNormalization())
```

#### Add two fully connected layers with 200 and 100 neurons respectively with `relu` activations. Add a dropout layer with `p=0.25`

```python
model.add(Dense(200, input_shape=(28,), activation='relu'))
model.add(Dense(100, activation='relu'))

# add dropout layer with p=0.25 after the 2 conv layers
model.add(tf.keras.layers.Dropout(0.25))
```

### Add the output layer with a fully connected layer with 10 neurons with `softmax` activation. Use `categorical_crossentropy` loss and `adam` optimizer and train the network. And, report the final validation.

```python
#Output layer
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# train the network
model.fit(x_train,train_Y,validation_data=(x_test,test_Y), epochs=10)
```

```python
# Evaluate the model on validation set
score= model.evaluate(x_test, test_Y, verbose=0)
# Print test accuracy
print('\n', 'Validation accuracy for the Model : {score_val}'.format(score_val=score[1]*100))
```
