```python
# 2.	Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)
```

```python
# Five Percent Significance Test
# Here NULL Hypothesis is that there is NO SIGNIFICANT Increase between the old scheme and the new scheme 
# ALTERNATE Hypothesis is that there is a significant Increase 
# We Do two-sample testing and check whether to reject Null Hypothesis or not.
```

```python
import numpy as np

old_sales = np.array(df["Old Scheme (in thousands)"])
new_sales = np.array(df["New Scheme (in thousands)"])
```

```python
from scipy.stats import ttest_ind
# two-sample t-test
# null hypothesis: There is no significant difference between data sets -->  mean (new) <= mean (old)
# Alternative Hypothesis that we are trying to prove  mean(new) > mean(old)
t_statistic, p_value = ttest_ind(old_sales, new_sales)
print("t statistic is " + str(t_statistic) + "\n" + "p_value is " + str(p_value))
```

```python
# 3.	What conclusion does the test (p-value) lead to? (2.5 points)
```

```python
# Answer 3 - This is a case of Right Tailed Test
# Here p-value > 0.05
# Means Null Hypothesis can NOT be Rejected 
# Means Mean of the New Sales figures are NOT Significantly Greater than the mean of the Old Sales figures
```

```python
# in order for Titan to break even, the average output must increase by £5000 in the scheme compared to the old scheme. 
```

```python
# The probability of a type 1 error ?
# Answer : As this is mentioned as a five percent significance test , the probability of a Type 1 error = 0.05
```

```python
# What is the p- value of the hypothesis test if we test for a difference of $5000? (5 points)
```

```python
# paired t-test: interested to know the difference between 2 measurements
# In this case Alternate Hypothesis is mean(new) - mean(old) > 5

from scipy.stats import ttest_1samp
t_pairedstatt, p_valuediff  = ttest_1samp(new_sales - old_sales, 5)
print("t statistic is " + str(t_pairedstatt))
# As this is a one-sides test of > 5000 and t-stat is -ve , pval = 1 - pval\2
p_valuediff = 1 - p_valuediff/2
print("pvalue : " + str(p_valuediff) )
```
