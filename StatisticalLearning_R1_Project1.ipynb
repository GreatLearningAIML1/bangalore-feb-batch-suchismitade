```python
# Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)
```

```python
import pandas as pd

df = pd.read_csv("Laliga.csv")
```

```python
df.head(10)
```

```python
for column in df:
    df[column] = df[column].replace("-",'0')
df.head(20)
```

```python
#Print all the teams which have started playing between 1930-1980. (5 points)
```

```python
df_team = []
for i in df.index:
    deb_yr = int(df.get_value(i,'Debut'))
    if(deb_yr >= 1930 and deb_yr <= 1980):
       df_team.append({'Team':df.get_value(i,'Team'), 'Debut': deb_yr})
df_team = pd.DataFrame(df_team)
df_team
```

```python
#Print the list of teams which came Top 5 in terms of points (2.5 points)
```

```python
df.sort_values('Points', ascending=False).head(5)[['Team', 'Points']]
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
```

```python
def Goal_diff_count(df1):
    df_team = []
    for i in df1.index:
        diff = int(df1.get_value(i,'GoalsFor')) - int(df1.get_value(i,'GoalsAgainst'))
        df_team.append({'Team':df1.get_value(i,'Team'), 'Diff': diff})
    return pd.DataFrame(df_team)
    

    
```

```python
df_ret = Goal_diff_count(df)
df_ret[["Diff"]] = df_ret[["Diff"]].apply(pd.to_numeric)
df_ret.sort_values(by='Diff',axis=0,ascending=False).tail(5)
```
